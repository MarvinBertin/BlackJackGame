```python
from BJdeck import Deck
from BJwar_player import Player


class BlackJack(object):
    def __init__(self):
        self.player = Player("Player")
        self.dealer = Player("Dealer")
        self.winner = None
        self.dollars = 50
        self.deck = Deck()
    
    def deal(self):
        self.deck = Deck()
        self.deck.shuffle()
        self.player.hand = []
        self.dealer.hand = []
        for _ in range(2):
            self.player.receive_card(self.deck.draw_card())
            self.dealer.receive_card(self.deck.draw_card())

    def play_game(self):
        while self.dollars > 0:
            gameOn = raw_input("Do you want to play a new game of Black Jack? (y/n): ")
            if gameOn == "y":
                self.deal()
                player = self.player
                while self.winner is None:
                    player = self.play_round(player)
                    if player == 'Over':
                        break

                self.display_winner()

                if self.winner == self.player.name:
                    self.dollars += 10
                else:
                    self.dollars -= 10     
                self.display_savings()
                self.winner = None
            else:
                break
        print "Game Over, you're bankrupt!"
                
    def display_savings(self):
        print "You have {}$ left.".format(self.dollars)
        
    def check_score(self, player):
        score = 0
        for card in player.hand:
            score += card.score()
        return score
    
    def hit(self):
        self.player.receive_card(self.deck.draw_card())
        return self.player
        
    def stand(self):
        score_player = self.check_score(self.player)
        score_dealer = self.check_score(self.dealer)
        if score_dealer > score_player:
            self.winner = self.dealer.name
            self.display_hand_score(self.dealer)
            return "Over"
        self.dealer.receive_card(self.deck.draw_card())
        return self.dealer
    
    def otherPlayer(self, name):
        if name == self.player.name:
            return self.dealer.name
        else:
            return self.player.name
        
    def display_hand_score(self, player):
        print "\n{} hand is: {}".format(player.name, player.hand)
        score = self.check_score(player)
        print "{} score is: {}".format(player.name, score)
    
    def play_round(self, player):
        self.display_hand_score(player)
        score = self.check_score(player)
        if self.check_ifOver(player, score):
            return "Over"
        
        if player == self.player:
            return self.players_turn(player)
        else:
            return self.dealers_turn(player, score)
        
    def players_turn(self, player):
        nextmove = raw_input("Do you want to hit or stand? (h/s): ") 

        if nextmove == "h":
            return self.hit()
        else:
            return self.stand()
    
    def dealers_turn(self, player, score):
        if score <= 16:
            self.dealer.receive_card(self.deck.draw_card())
            return self.dealer
        else:
            score_player = self.check_score(self.player)
            if score > score_player:
                self.winner = self.dealer.name
            else:
                self.winner = self.player.name
            return "Over"
            
    def check_ifOver(self, player, score):
        if score == 21:
            self.winner = player.name
            return True
        elif score > 21:
            self.winner = self.otherPlayer(player.name)
            print "Busted!"
            return True        
        
    def cards_to_str(self, cards):
        return " ".join(str(card) for card in cards)
    
    def display_winner(self):
        print "The winner is %s!!!" % self.winner
        
if __name__ == '__main__':
    game = BlackJack()
    game.play_game()
```
